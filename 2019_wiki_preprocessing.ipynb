```python
import numpy as np
import pandas as pd
```

```python
link = 'https://en.wikipedia.org/wiki/List_of_American_films_of_2019'
df1 = pd.read_html(link)[2]
df2 = pd.read_html(link)[3]
df3 = pd.read_html(link)[4]
df4 = pd.read_html(link)[5]
```

```python
print(df1.columns)
print(df2.columns)
print(df3.columns)
print(df4.columns)
```

```python
df1.drop(columns=['Opening', 'Opening.1','Production company','Ref.'], inplace=True)
df2.drop(columns=['Opening', 'Opening.1','Production company','Ref.'], inplace=True)
df3.drop(columns=['Opening', 'Opening.1','Production company','Ref.'], inplace=True)
df4.drop(columns=['Opening', 'Opening.1','Production company','Ref.'], inplace=True)
```

```python
df1.rename({"Title":"movie_title", "Cast and crew": 'cast_and_crew'}, axis=1, inplace=True)
df2.rename({"Title":"movie_title", "Cast and crew": 'cast_and_crew'}, axis=1, inplace=True)
df3.rename({"Title":"movie_title", "Cast and crew": 'cast_and_crew'}, axis=1, inplace=True)
df4.rename({"Title":"movie_title", "Cast and crew": 'cast_and_crew'}, axis=1, inplace=True)
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('max_colwidth', -1)
```

```python
final_df = pd.concat([df1,df2,df3,df4], ignore_index=True, sort=False)
final_df = final_df.drop(index=[33,83,89]).reset_index()
final_df
```

```python
def get_director(x):
    if '(director)' in x:
        return x.split("(director)")[0]
    elif "(directors)" in x:
        return x.split("(directors)")[0]
    elif "(director/screenplay)" in x:
        return x.split("(director/screenplay)")[0]
    elif "(directors/screenplay)" in x:
        return x.split("(directors/screenplay)")[0]
    
final_df['director_name'] = final_df['cast_and_crew'].map(lambda x: get_director(x))    
```

```python
final_df = final_df.drop(index=[13,175]).reset_index()


def get_actor_1(x):
    if '(screenplay)' in x:
        return ((x.split("(screenplay); ")[-1]).split(',')[0])
    elif "(co-director/screenplay)" in x:
        return ((x.split("(co-director/screenplay); ")[-1]).split(',')[0])
    else:
        return ((x.split(";")[-1]).split(', ')[0])

final_df['actor_1_name'] = final_df['cast_and_crew'].map(lambda x: get_actor_1(x))

```

```python
final_df.drop(columns=['level_0','index'],inplace=True)
final_df = final_df.drop(index=[61,146]).reset_index()


def get_actor2(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])
    
final_df['actor_2_name'] = final_df['cast_and_crew'].apply(get_actor2)
```

```python
def get_actor3(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
           return np.NaN
    else:
           return ((x.split("screenplay); ")[-1]).split(",")[2])

        
final_df['actor_3_name'] = final_df['cast_and_crew'].apply(get_actor3)    
```

```python
final_df.drop(columns=['index'], inplace=True)
```

```python
movie = Movie()
def get_genre(x):
    genres = []
    result = movie.search(x)
    #print(result)
    movie_id = result[0].id
    #print(movie_id)
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key=10c4c258c4c44208fe2be289f962bd9e'.format(movie_id))
    data_json = response.json()
    #print(data_json['genres'])
    if data_json['genres']:
        genre_str = " "
        for i in range(0, len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name'])
        genres_list = str(genres) 
        return genres_list
    else:
        return np.NaN
#get_genre('Replicas')
final_df['Genre'] = final_df['movie_title'].map(lambda x: get_genre(str(x)))
```

```python
def list_to_string(x):
    return x.replace(',', '').replace('[','').replace(']','').replace("'", '')
final_df['genres'] = final_df["Genre"].map(lambda x: list_to_string(x))
final_df.drop(columns="Genre", inplace=True)
final_df
```
